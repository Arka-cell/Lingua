```python
import Lingua # Lingua module provides classes to structure data for statistical analysis.
```

```python
text = "Also a pioneer in online education, Ng co-founded Coursera and deeplearning.ai"
```

```python
ling = Lingua.Linguistics(text) # Creating a lingua object. Sentences were tokenized by spaCy sentence tokenizer.
```

```python
print(ling.sentences) # Returning sentences as spacy.tokens.doc.Doc object.
```

```python
ling.lengths() # Returning sentence lengths for sentences in lingua.
```

```python
ling.deplengths(punc=False) # Returning dependency lengths for each sentence. Because dependecy length algorithm is based on 
                            # spacy token.children, punctuations will be considered as children and inflate lengths.
                            # To avoid this, use punc=True.
```

```python
ling.deplengths(punc=True) # Dependency lengths without punctuation. The results went from 30 and 86 to 22 and 53, respectively.
```

```python
ling.getSentence(0,2) # Return sentences based on indexing, end_index is optional. Negative indexing is available.
```

```python
ling.getSentence(-1)
```

```python

ling.tree(0) # Return dependency tree using displaCy.render.
```

```python
ling.pos().head() # Returns a pandas dataframe with row representing each word part-of-speech in a sentence. 
```

```python
ling.dep().head() # Returns a pandas dataframe with row representing each word dependecy in a sentence. 
```

```python
ling.words().head() # Returns a pandas dataframe with row representing each word text in a sentence. 
```

```python
ling.postag().head() # Returns a pandas dataframe with row representing each word tag in a sentence. 
```
